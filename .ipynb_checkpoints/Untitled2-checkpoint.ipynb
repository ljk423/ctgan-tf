```python
import tensorflow as tf
import torch
import numpy as np
```

```python
data = np.random.rand(10, 5)
data
```

```python
nb_classes = 5
targets = np.array([[np.random.randint(0, nb_classes) for i in range(10)]]).reshape(-1)
one_hot_targets = np.eye(nb_classes)[targets]
one_hot_targets
```

```python
c = np.array([np.random.randint(0, nb_classes) for i in range(10)])
```

```python
tf.nn.sparse_softmax_cross_entropy_with_logits(tf.convert_to_tensor(c), tf.convert_to_tensor(data))
```

```python
torch.nn.functional.cross_entropy(torch.from_numpy(data), torch.from_numpy(c), reduction='none')
```
